10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python ?

    -->The map function applies a given function to all items in an input iterable (like a list) and returns an iterator with the results.

    The reduce function from the functools module applies a given function cumulatively to the items of a sequence, from left to right, to reduce the sequence to a single value.

    The filter function constructs an iterator from elements of an iterable for which a specified function returns True.

```python
# 1.  Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.
def sum_even_no(a):
    sum=0
    for i in a:
        if i%2==0:
            sum=sum+i
    return sum
sum_even_no([1,44,556])
```

```python
# 3.  Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.
def sq_no(n):
  result=[]
  for i in n:
    result.append(i**2)
  return result
sq_no([4,66,44])
```

```python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True
for i in range(1,200):
    if is_prime(i):
        print(i)



```

```python
# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

class FibonacciItrator:
  def __init__(self,n):
    self.n = n
    self.a,self.b = 0,1

  def __iter__(self):
    return self

  def __next__(self):
    if self.n<=0:
      raise StopIteration
    self.n -= 1
    self.a,self.b = self.b,self.a+self.b
    return self.a

fiibonacci = FibonacciItrator(10)
for num in fiibonacci:
  print(num)
```

```python
# 6.  Write a generator function in Python that yields the powers of 2 up to a given exponent.
def a(exponent):
  for i in range(exponent+1):
    yield 2 ** i

for i in a(5):
  print(i)

```

```python
# 7. Implement a generator function that reads a file line by line and yields each line as a string.
def read_file_line_by_line(filename):
    with open(filename, 'r') as file:
        for line in file:
            yield line.strip()  # strip() removes the trailing newline characters

# Example usage:
filename = ''  # Replace with your actual file path
for line in read_file_line_by_line(filename):
    print(line)
```

```python
# 9.  Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

def c_to_f(Celsius):
  return (Celsius * 9/5) + 32

celsious_temprature = [66,43,56,332,43,53,6,4]

farenheit_temratiue = list(map(c_to_f,celsious_temprature))
print(farenheit_temratiue)
```

```python
# 10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

def a(char):
  return char.lower() not in "aeiou"

b = "Hello World!"

c = "".join(filter(a,b))
print(c)
```
